```python
df['discounted_price'] = df['discounted_price'].str.lstrip('â‚¹').str.replace(",","").astype('float64')
```

```python
# now working on discount_percentage
df['discount_percentage'].sample(30)
```

```python
# now working on rating
df['rating'] = df['rating'].astype(str).str.replace('|',"0").astype('float32')
```

```python
# now working on rating count
df['rating_count'] = df['rating_count'].fillna('0').astype(str).str.replace(",","").astype('float64')
```

```python
df['rating_count'] = df['rating_count'].replace(0,df['rating_count'].mean())
```
